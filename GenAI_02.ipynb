```python
from scipy.optimize import linprog
import math as mt
import numpy as np
```

```python
obj1 = [-6,-5]
A1 = [[1,1],[3,2]]
B1 = [5,12]
bnd = [(0,float("inf")),(0,float("inf"))]
opt = linprog(c=obj1,A_ub=A1,b_ub=B1,bounds=bnd)
choc=np.round(opt.x)
print(f"No. of A : {int(choc[0])}")
print(f"No. of B : {int(choc[1])}")
```
